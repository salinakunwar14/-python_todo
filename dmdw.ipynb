notna(), notnull() and using any() or sum() functions individually with each of them.

```python
import pandas as pd
import numpy as np

# Step 1: Create Dataset
data = {
    "CustomerID": ["C001","C002","C003","C004","C004",
                   "C005","C006","C007","C008","C008"],

    "Name": ["John","Alice","BOB","BOBY","Eve",
             "eve","Steve","Ramu","mary","Bob"],

    "Age": [25,34,17,29,29,
            120,-5,np.nan,220,30],

    "JoinDate": ["12/1/2024","11/15/2023","6/1/2022",
                 "6/1/2022","12/5/2024",
                 "invalid_date","",
                 "1/1/2024","3/5/2023","3/5/2023"],

    "MonthlyCharges": [29.85,56.95,4000,75.5,75.5,
                       45.99,60,49.99,-30,55],

    "Churn": ["No","Yes","No","No","No",
              "Yes","No",np.nan,"Yes","No"]
}

# Step 2: Convert into DataFrame
df = pd.DataFrame(data)

# Step 3: Display Dataset
print("Dataset")
print(df)

# Step 4: isnull()
print("\nUsing isnull()")
print(df.isnull())

# Step 5: isna()
print("\nUsing isna()")
print(df.isna())

# Step 6: notnull()
print("\nUsing notnull()")
print(df.notnull())

# Step 7: notna()
print("\nUsing notna()")
print(df.notna())

# Step 8: sum()
print("\nMissing Values Count")
print(df.isnull().sum())

# Step 9: any()
print("\nCheck Missing Values using any()")
print(df.isnull().any())
```

```python
import pandas as pd

# Step 1: Create Dataset
data = {
    "CustomerID": ["C001","C002","C003","C004","C004",
                   "C005","C006","C007","C008","C008"],

    "Name": ["John","Alice","BOB","BOBY","Eve",
             "eve","Steve","Ramu","mary","Bob"],

    "Age": [25,34,17,29,29,
            120,-5,None,220,30],

    "JoinDate": ["12/1/2024","11/15/2023","6/1/2022",
                 "6/1/2022","12/5/2024",
                 "invalid_date","",
                 "1/1/2024","3/5/2023","3/5/2023"],

    "MonthlyCharges": [29.85,56.95,4000,75.5,75.5,
                       45.99,60,49.99,-30,55],

    "Churn": ["No","Yes","No","No","No",
              "Yes","No",None,"Yes","No"]
}

# Step 2: Convert into DataFrame
df = pd.DataFrame(data)

# Step 3: Display Original Dataset
print("Original Dataset")
print(df)

# Step 4: Find Duplicate Rows
duplicates = df[df.duplicated(subset=["CustomerID"])]

print("\nDuplicate Rows")
print(duplicates)

# Step 5: Remove Duplicate Rows
df_clean = df.drop_duplicates(subset=["CustomerID"])

print("\nDataset after Removing Duplicates")
print(df_clean)
```

min max normalization

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# Step 1: Create Dataset
data = {
    "Feature1": [10, 20, 30, 40, 50],
    "Feature2": [100, 200, 300, 400, 500],
    "Feature3": [5, 15, 25, 35, 45]
}

# Step 2: Convert into DataFrame
df = pd.DataFrame(data)

# Step 3: Display Original Dataset
print("Original Dataset")
print(df)

# -----------------------------------
# a) Min-Max Normalization
# -----------------------------------

minmax_scaler = MinMaxScaler()

minmax_data = minmax_scaler.fit_transform(df)

print("\nMin-Max Normalization")
print(minmax_data)

# -----------------------------------
# b) Z-Score Normalization
# -----------------------------------

zscore_scaler = StandardScaler()

zscore_data = zscore_scaler.fit_transform(df)

print("\nZ-Score Normalization")
print(zscore_data)

# -----------------------------------
# c) Decimal Scale Normalization
# -----------------------------------

decimal_scaled = df / 100

print("\nDecimal Scale Normalization")
print(decimal_scaled)
```

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Step 1: Create Dataset
data = {
    "BMI": [18.5, 24.0, 30.2, 35.6, 28.1],
    "Age": [20, 25, 30, 45, 35],
    "Risk": ["Low", "Low", "Medium", "High", "Medium"]
}

# Step 2: Convert into DataFrame
df = pd.DataFrame(data)

# Step 3: Display Dataset
print("Dataset")
print(df)

# Step 4: Define Features and Target
X = df[["BMI", "Age"]]

y = df["Risk"]

# Step 5: Create KNN Model
model = KNeighborsClassifier(n_neighbors=3)

# Step 6: Train Model
model.fit(X, y)

# Step 7: New Data Point
new_point = [[43.6, 40]]

# Step 8: Predict
prediction = model.predict(new_point)

# Step 9: Display Result
print("\nPredicted Class:")
print(prediction)
```

```python
import pandas as pd

# Step 1: Create Dataset
data = {
    "Staff": [101, 102, 103, 104, 105],
    "Age": [5, 3, 2, 4, 7],
    "Weight": [20, 8, 0.5, 18, 9],
    "Color": ["Brown", "Black", "Yellow", "White", "Brown"]
}

# Step 2: Convert into DataFrame
df = pd.DataFrame(data)

# Step 3: Display Original Dataset
print("Original Dataset")
print(df)

# Step 4: Apply One-Hot Encoding
encoded = pd.get_dummies(df["Color"])

# Step 5: Display Encoded Features
print("\nEncoded Features")
print(encoded)

# Step 6: Combine Dataset
result = pd.concat([df, encoded], axis=1)

# Step 7: Display Final DataFrame
print("\nFinal DataFrame")
print(result)
```

Detect Outlier using IQR Method and Plot Boxplot

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create Dataset
data = [50, 55, 52, 48, 51, 60, 62, 54, 53, 47, 65, 90]

# Step 2: Display Dataset
print("Dataset")
print(data)

# Step 3: Find Q1
Q1 = np.percentile(data, 25)

print("Q1 =", Q1)

# Step 4: Find Q3
Q3 = np.percentile(data, 75)

print("Q3 =", Q3)

# Step 5: Calculate IQR
IQR = Q3 - Q1

print("IQR =", IQR)

# Step 6: Calculate Lower Limit
lower_limit = Q1 - 1.5 * IQR

print("Lower Limit =", lower_limit)

# Step 7: Calculate Upper Limit
upper_limit = Q3 + 1.5 * IQR

print("Upper Limit =", upper_limit)

# Step 8: Detect Outliers
outliers = []

for x in data:
    if x < lower_limit or x > upper_limit:
        outliers.append(x)

print("Outliers =", outliers)

# Step 9: Draw Boxplot
plt.boxplot(data)

plt.title("Box Plot")

plt.show()
```
